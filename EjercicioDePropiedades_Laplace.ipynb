**Ejemplo - Transformada de una se√±al coseno**

La transformada de Laplace de una se√±al coseno causal se calcula con `Sympy`

```python
%matplotlib inline
import sympy as sym
sym.init_printing()
#se definen las variables simb√≥licas y se aplica la funci√≥n de laplace de sympy
t, w0 = sym.symbols('t omega0', real=True)
s = sym.symbols('s', complex=True)

x = sym.sin(w0*t)
X, a, cond = sym.laplace_transform(x, t, s)
X, a, cond
```

# üìù Ejercicio - Polos y ceros de $\mathcal{L}\{u(t)\sin(\omega_0 t)\}$

---

## 1. üßÆ Transformada de Laplace

Dado que $u(t)$ es la **funci√≥n escal√≥n unitario** (se√±al causal), la transformada de Laplace de $u(t)\sin(\omega_0 t)$ es:

$$\mathcal{L}\{u(t)\sin(\omega_0 t)\} = \frac{\omega_0}{s^2 + \omega_0^2}, \quad \text{Re}(s) > 0.$$

---

## 2. üìç Polos y ceros

Expresamos la funci√≥n de transferencia $X(s)$ con el denominador en forma factorizada:

$$X(s) = \frac{\omega_0}{s^2 + \omega_0^2} = \frac{\omega_0}{(s - j\omega_0)(s + j\omega_0)}.$$

* **Ceros finitos:**
    * El numerador es una constante $\omega_0 \neq 0$.
    * Por lo tanto, **no hay ceros finitos** en el plano $s$. (Se dice que hay un cero en el infinito).

* **Polos:**
    * Se obtienen de los valores que anulan el denominador: $s^2 + \omega_0^2 = 0 \Rightarrow s = \pm j\omega_0$.
    * **Polos:** $p_1 = j\omega_0$, $p_2 = -j\omega_0$.

---

## 3. ‚ú® Simetr√≠a en el plano $s$

* Los dos polos $j\omega_0$ y $-j\omega_0$ son **complejos conjugados**, sim√©tricos respecto al eje real.
* Esta simetr√≠a conjugada es caracter√≠stica de las transformadas de Laplace de **se√±ales reales en el tiempo** (como $u(t)\sin(\omega_0 t)$).
* En este caso, los polos est√°n sobre el **eje imaginario**, lo que corresponde a una se√±al sinusoidal pura y no amortiguada (sin decaimiento exponencial).

---

## 4. üîÑ Comparaci√≥n con $u(t)\cos(\omega_0 t)$

Para completar el an√°lisis, la transformada de Laplace del coseno es:

$$\mathcal{L}\{u(t)\cos(\omega_0 t)\} = \frac{s}{s^2 + \omega_0^2}.$$

* **Polos:** $s = \pm j\omega_0$ (misma ubicaci√≥n que en el seno).
* **Cero finito:** $s = 0$ (sobre el eje real).
* La simetr√≠a conjugada de los polos se mantiene.

---

## 5. üó∫Ô∏è Diagrama de polos y ceros (cualitativo)

En el plano $s$ ($\sigma$ es el eje real y $j\omega$ es el eje imaginario):

Para $X(s) = \frac{\omega_0}{s^2 + \omega_0^2}$:

* Polos ($\times$): marcados en $s = j\omega_0$ y $s = -j\omega_0$ (sobre el eje imaginario).
* Ceros ($\circ$): ninguno finito.
