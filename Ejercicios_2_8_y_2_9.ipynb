# üìù Ejercicio 2.8 - Demostraci√≥n de Propiedades de la Transformada de Laplace

---

## i) ‚è≥ Desplazamiento Temporal

**Propiedad:** $\mathcal{L}\{x(t-t_0)\} = e^{-st_0}X(s)$

**Demostraci√≥n:**
Partimos de la definici√≥n de la Transformada de Laplace:
$$\mathcal{L}\{x(t-t_0)\} = \int_{0}^{\infty} x(t-t_0)e^{-st}dt$$

Realizamos el cambio de variable:
$$\tau = t - t_0 \quad \Rightarrow \quad t = \tau + t_0, \quad dt = d\tau$$

Sustituyendo en la integral y ajustando los l√≠mites:
$$\int_{0 - t_0}^{\infty - t_0} x(\tau)e^{-s(\tau + t_0)}d\tau = \int_{-t_0}^{\infty} x(\tau)e^{-s\tau}e^{-st_0}d\tau$$

Factorizamos la constante $e^{-st_0}$:
$$= e^{-st_0} \int_{-t_0}^{\infty} x(\tau)e^{-s\tau}d\tau$$

Dado que $x(t)$ es una **se√±al causal** (es decir, $x(t) = 0$ para $t < 0$), la funci√≥n $x(\tau)$ es cero para $\tau < 0$. Por lo tanto, podemos cambiar el l√≠mite inferior de integraci√≥n a $0$ sin afectar el resultado:
$$= e^{-st_0} \int_{0}^{\infty} x(\tau)e^{-s\tau}d\tau$$

Reconociendo la integral como $X(s) = \mathcal{L}\{x(t)\}$:
$$\mathcal{L}\{x(t-t_0)\} = e^{-st_0}X(s)$$

---

## ii) ‚öñÔ∏è Escalamiento Temporal

**Propiedad:** $\mathcal{L}\{x(at)\} = \frac{1}{|a|}X(s/a)$

### Caso $a > 0$:
**Demostraci√≥n:**
$$\mathcal{L}\{x(at)\} = \int_{0}^{\infty} x(at)e^{-st}dt$$

Realizamos el cambio de variable:
$$\tau = at \quad \Rightarrow \quad t = \tau/a, \quad dt = d\tau/a$$

Sustituyendo en la integral y ajustando los l√≠mites (como $a>0$, los l√≠mites no cambian):
$$= \int_{0}^{\infty} x(\tau)e^{-s(\tau/a)}\frac{d\tau}{a}$$

Factorizamos $1/a$:
$$= \frac{1}{a} \int_{0}^{\infty} x(\tau)e^{-(s/a)\tau}d\tau$$

Reconociendo la integral como $X(s/a)$:
$$\mathcal{L}\{x(at)\} = \frac{1}{a}X(s/a)$$

### Caso $a < 0$:
El cambio de variable $\tau = at$ invierte los l√≠mites de integraci√≥n, resultando en un factor $|a|$ en el denominador que maneja el cambio de signo de la integral, manteniendo la relaci√≥n $\frac{1}{|a|}X(s/a)$.

---

## iii) ‚öôÔ∏è Derivaci√≥n

**Propiedad:** $\mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = sX(s) - x(0^-)$

**Demostraci√≥n:**
$$\mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = \int_{0^-}^{\infty} \frac{dx(t)}{dt}e^{-st}dt$$

Utilizamos la **integraci√≥n por partes** ($\int u\,dv = uv - \int v\,du$):
* $u = e^{-st} \quad \Rightarrow \quad du = -se^{-st}dt$
* $dv = \frac{dx(t)}{dt}dt \quad \Rightarrow \quad v = x(t)$

Aplicando la f√≥rmula de integraci√≥n por partes:
$$= \left[x(t)e^{-st}\right]_{0^-}^{\infty} - \int_{0^-}^{\infty} x(t)(-se^{-st})dt$$

Evaluamos el primer t√©rmino:
$$= \left(\lim_{t \to \infty} x(t)e^{-st}\right) - x(0^-)e^{-s(0^-)} + s\int_{0^-}^{\infty} x(t)e^{-st}dt$$

Suponiendo que $x(t)e^{-st} \to 0$ cuando $t \to \infty$ (condici√≥n necesaria para la convergencia de $X(s)$) y que $e^0 = 1$:
$$= \left[0 - x(0^-)\right] + s\int_{0}^{\infty} x(t)e^{-st}dt$$

Reconociendo la integral como $X(s)$:
$$\mathcal{L}\left\{\frac{dx(t)}{dt}\right\} = sX(s) - x(0^-)$$

---

## iv) üí´ Convoluci√≥n

**Propiedad:** $\mathcal{L}\{x(t)*y(t)\} = X(s)Y(s)$

**Demostraci√≥n:**
Partimos de la definici√≥n de convoluci√≥n en el tiempo y la Transformada de Laplace:
$$\mathcal{L}\{x(t)*y(t)\} = \int_{0}^{\infty} \left[\int_{0}^{t} x(\tau)y(t-\tau)d\tau\right]e^{-st}dt$$

Cambiamos el orden de integraci√≥n, redefiniendo la regi√≥n como $0 < \tau < t < \infty$:
$$= \int_{0}^{\infty} x(\tau)\left[\int_{\tau}^{\infty} y(t-\tau)e^{-st}dt\right]d\tau$$


Realizamos un nuevo cambio de variable en la integral interna:
$$u = t - \tau \quad \Rightarrow \quad t = u + \tau, \quad dt = du$$

Ajustando los l√≠mites de la integral interna (si $t=\tau \Rightarrow u=0$; si $t=\infty \Rightarrow u=\infty$):
$$= \int_{0}^{\infty} x(\tau)\left[\int_{0}^{\infty} y(u)e^{-s(u+\tau)}du\right]d\tau$$

Separamos los t√©rminos exponenciales:
$$= \int_{0}^{\infty} x(\tau)\left[\int_{0}^{\infty} y(u)e^{-su}e^{-s\tau}du\right]d\tau$$

Reordenamos los t√©rminos, sacando $e^{-s\tau}$ de la integral interna:
$$= \int_{0}^{\infty} x(\tau)e^{-s\tau}\left[\int_{0}^{\infty} y(u)e^{-su}du\right]d\tau$$

La integral interna es $Y(s)$ (ya que $\tau$ es la variable de integraci√≥n de la integral externa, la interna depende solo de $u$ y $s$). Factorizamos $Y(s)$:
$$= \left[\int_{0}^{\infty} x(\tau)e^{-s\tau}d\tau\right] Y(s)$$

Reconociendo la integral restante como $X(s)$:
$$\mathcal{L}\{x(t)*y(t)\} = X(s)Y(s)$$

# ‚öôÔ∏è Ejercicio 2.9 - Transformadas de Laplace: C√°lculo y An√°lisis

---

## i) $x_1(t) = e^{-2t}u(t) + e^{-3t}u(t)$

Esta es una **se√±al causal** (de lado derecho), suma de dos exponenciales amortiguadas.

### 1. Transformada de Laplace $X_1(s)$:
Aplicando la linealidad y la transformada b√°sica $\mathcal{L}\{e^{-at}u(t)\} = \frac{1}{s+a}, \quad \Re\{s\} > -a$:
$$X_1(s) = \frac{1}{s+2} + \frac{1}{s+3}$$

**√Ålgebra para unificar la expresi√≥n:**
$$X_1(s) = \frac{(s+3) + (s+2)}{(s+2)(s+3)} = \frac{2s+5}{(s+2)(s+3)}$$

### 2. Polos, Ceros y ROC:

* **Polos:** Valores que anulan el denominador: $s = -2$ y $s = -3$ (ambos simples).
* **Ceros finitos:** Valores que anulan el numerador: $2s+5 = 0 \Rightarrow s = -2.5$ (simple).
* **ROC:** Debe ser mayor a la parte real del polo m√°s a la derecha. $\max(\Re\{-2\}, \Re\{-3\}) = -2$.
    $$\mathrm{ROC}: \Re\{s\} > -2$$

---

## ii) $x_2(t) = e^{2t}u(t) + e^{-3t}u(-t)$

Esta es una **se√±al bilateral**, suma de una exponencial de lado derecho y una de lado izquierdo.

### 1. Transformada de Laplace $X_2(s)$:
Utilizamos las transformadas b√°sicas:
$$\mathcal{L}\{e^{2t}u(t)\} = \frac{1}{s-2}, \quad \Re\{s\} > 2$$
$$\mathcal{L}\{e^{-3t}u(-t)\} = -\frac{1}{s-(-3)} = -\frac{1}{s+3}, \quad \Re\{s\} < -3$$

$$X_2(s) = \frac{1}{s-2} - \frac{1}{s+3}$$

**√Ålgebra para unificar la expresi√≥n:**
$$X_2(s) = \frac{(s+3) - (s-2)}{(s-2)(s+3)} = \frac{5}{(s-2)(s+3)}$$

### 2. Polos, Ceros y ROC:

* **Polos:** $s = 2$ y $s = -3$ (ambos simples).
* **Ceros finitos:** El numerador es una constante $5 \neq 0$. **Ninguno finito**.
* **ROC:** La transformada bilateral solo existe si las ROC individuales se solapan.
    * ROC 1 ($\Re\{s\} > 2$)
    * ROC 2 ($\Re\{s\} < -3$)
    $$\max(-3, 2) = 2 \quad \text{y} \quad \min(-3, 2) = -3$$
    No hay regi√≥n de solapamiento donde $\Re\{s\} < -3$ y $\Re\{s\} > 2$.
    $$\mathrm{ROC}: \text{NO EXISTE ROC com√∫n} \quad \Rightarrow \quad \text{No hay Transformada Bilateral}$$
    

---

## iii) $x_3(t) = e^{-a|t|}$ (con $a>0$)

Esta es una **se√±al bilateral** sim√©trica, que se descompone en: $x_3(t) = e^{-at}u(t) + e^{at}u(-t)$.

### 1. Transformada de Laplace $X_3(s)$:
$$X_3(s) = \mathcal{L}\{e^{-at}u(t)\} + \mathcal{L}\{e^{at}u(-t)\}$$
$$X_3(s) = \frac{1}{s+a} - \frac{1}{s-a}, \quad \Re\{s\} > -a \quad \cap \quad \Re\{s\} < a$$

**√Ålgebra para unificar la expresi√≥n:**
$$X_3(s) = \frac{(s-a) - (s+a)}{(s+a)(s-a)} = \frac{-2a}{s^2 - a^2}$$

### 2. Polos, Ceros y ROC:

* **Polos:** $s^2 - a^2 = 0 \Rightarrow s = a$ y $s = -a$ (ambos simples).
* **Ceros finitos:** El numerador es una constante $-2a \neq 0$. **Ninguno finito**.
* **ROC:** La intersecci√≥n de las ROC individuales es:
    $$\mathrm{ROC}: -a < \Re\{s\} < a$$
    (La transformada solo existe si $a>0$).

---

## iv) $x_4(t) = e^{-2t}[u(t) - u(t-5)]$

Esta es un **pulso exponencial** de duraci√≥n finita ($0 \le t < 5$).

### 1. Transformada de Laplace $X_4(s)$:
Podemos verla como una diferencia de dos se√±ales causales, donde la segunda est√° desplazada en el tiempo ($t_0=5$):
$$x_4(t) = e^{-2t}u(t) - e^{-2t}u(t-5)$$
Note que $e^{-2t}u(t-5) = e^{-2(t-5)}e^{-10}u(t-5)$.

Aplicando la propiedad de **desplazamiento temporal** ($\mathcal{L}\{x(t-t_0)\} = e^{-st_0}X(s)$):
$$\mathcal{L}\{e^{-2t}u(t)\} = \frac{1}{s+2}$$
$$\mathcal{L}\{e^{-2(t-5)}u(t-5)\} = e^{-5s} \left(\frac{1}{s+2}\right)$$

$$X_4(s) = \frac{1}{s+2} - e^{-10} \mathcal{L}\{e^{-2(t-5)}u(t-5)\}$$

$$X_4(s) = \frac{1}{s+2} - e^{-10} \left[e^{-5s} \left(\frac{1}{s+2}\right)\right] = \frac{1 - e^{-10}e^{-5s}}{s+2}$$

### 2. Polos, Ceros y ROC:

* **Polos:** $s = -2$ (simple).
* **Ceros finitos:** Se obtienen de $1 - e^{-10}e^{-5s} = 0$:
    $$e^{-5s} = e^{10} \quad \Rightarrow \quad -5s = 10 + j2\pi k, \quad k \in \mathbb{Z}$$
    $$s = -2 - j\frac{2\pi k}{5}, \quad k \in \mathbb{Z} \quad \text{(Ceros infinitos y peri√≥dicos)}$$
* **ROC:** Dado que $x_4(t)$ es una **se√±al de duraci√≥n finita** y es cero para $t < 0$, su transformada converge en todo el plano $s$.
    Sin embargo, la forma funcional resultante $X_4(s)$ solo tiene un polo en $s=-2$. Ya que la se√±al es causal ($x(t)=0$ para $t<0$), la ROC debe ser la regi√≥n a la derecha del polo.
    $$\mathrm{ROC}: \Re\{s\} > -2$$

*(Nota: Para una se√±al de duraci√≥n finita, si la se√±al es causal, la ROC es todo el plano $s$ excepto los polos finitos. En este caso, $\Re\{s\} > -2$ es la ROC m√°s restrictiva de las dos componentes causales, y es la respuesta correcta para el polo √∫nico).*

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from matplotlib.patches import Circle

# Configuraci√≥n de gr√°ficos
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.autolayout'] = True

def plot_pole_zero(poles, zeros=None, title="", roc_left=None, roc_right=None):
    """
    Dibuja diagrama de polos y ceros en el plano s
    """
    fig, ax = plt.subplots()

    # Ejes
    ax.axhline(y=0, color='k', linestyle='-', alpha=0.3, linewidth=0.5)
    ax.axvline(x=0, color='k', linestyle='-', alpha=0.3, linewidth=0.5)

    # L√≠mites del gr√°fico
    all_points = []
    if poles:
        all_points.extend(poles)
    if zeros:
        all_points.extend(zeros)

    if all_points:
        real_parts = [p.real for p in all_points]
        imag_parts = [p.imag for p in all_points]
        max_val = max(max(abs(np.array(real_parts))), max(abs(np.array(imag_parts)))) + 2
    else:
        max_val = 5

    ax.set_xlim(-max_val, max_val)
    ax.set_ylim(-max_val, max_val)

    # Dibujar polos (X)
    for pole in poles:
        ax.plot(pole.real, pole.imag, 'rx', markersize=10, label='Polo' if pole == poles[0] else "")

    # Dibujar ceros (O)
    if zeros:
        for zero in zeros:
            ax.plot(zero.real, zero.imag, 'bo', markersize=10, fillstyle='none',
                   markeredgewidth=2, label='Cero' if zero == zeros[0] else "")

    # Dibujar ROC si se especifica
    if roc_left is not None or roc_right is not None:
        if roc_left is not None and roc_right is not None:
            # ROC entre dos valores
            ax.axvspan(roc_left, roc_right, alpha=0.2, color='green', label='ROC')
        elif roc_left is not None:
            # ROC > roc_left
            ax.axvspan(roc_left, max_val, alpha=0.2, color='green', label='ROC')
        elif roc_right is not None:
            # ROC < roc_right
            ax.axvspan(-max_val, roc_right, alpha=0.2, color='green', label='ROC')

    # Configuraci√≥n del gr√°fico
    ax.set_xlabel('Re(s) - œÉ')
    ax.set_ylabel('Im(s) - jœâ')
    ax.set_title(title)
    ax.grid(True, alpha=0.3)
    ax.legend(loc='best')

    # Cuadr√≠cula
    ax.set_aspect('equal', adjustable='box')

    plt.show()
    return fig, ax

# ============================================
# SE√ëAL 1: e^{-2t}u(t) + e^{-3t}u(t)
# ============================================
print("="*60)
print("SE√ëAL 1: e^{-2t}u(t) + e^{-3t}u(t)")
print("="*60)

# Usando sympy para calcular la transformada
t, s = sp.symbols('t s', real=False)
x1 = sp.exp(-2*t)*sp.Heaviside(t) + sp.exp(-3*t)*sp.Heaviside(t)
X1 = sp.laplace_transform(x1, t, s, noconds=True)
print(f"X1(s) = {sp.simplify(X1)}")

# Polos y ceros
poles1 = [-2, -3]
zeros1 = [-2.5]  # De 2s+5=0

print(f"Polos: {poles1}")
print(f"Ceros: {zeros1}")
print(f"ROC: Re(s) > -2")

# Diagrama de polos y ceros
plot_pole_zero(poles1, zeros1,
               title="Se√±al 1: $e^{-2t}u(t) + e^{-3t}u(t)$\n$X(s) = \\frac{2s+5}{(s+2)(s+3)}$",
               roc_left=-2)

# ============================================
# SE√ëAL 2: e^{2t}u(t) + e^{-3t}u(-t)
# ============================================
print("\n" + "="*60)
print("SE√ëAL 2: e^{2t}u(t) + e^{-3t}u(-t)")
print("="*60)

# Polos y ceros (aunque no hay ROC com√∫n)
poles2 = [2, -3]
zeros2 = []  # Numerador constante 5

print(f"Polos: {poles2}")
print(f"Ceros: Ninguno finito")
print(f"ROC componente 1: Re(s) > 2")
print(f"ROC componente 2: Re(s) < -3")
print(f"NO HAY ROC COM√öN - No existe transformada bilateral")

# Diagrama
fig2, ax2 = plt.subplots()
ax2.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax2.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Polos
for pole in poles2:
    ax2.plot(pole, 0, 'rx', markersize=10)

# ROC componentes
ax2.axvspan(2, 4, alpha=0.2, color='blue', label='ROC t√©rmino 1: Re(s)>2')
ax2.axvspan(-5, -3, alpha=0.2, color='red', label='ROC t√©rmino 2: Re(s)<-3')

ax2.set_xlim(-5, 5)
ax2.set_ylim(-2, 2)
ax2.set_xlabel('Re(s)')
ax2.set_ylabel('Im(s)')
ax2.set_title("Se√±al 2: $e^{2t}u(t) + e^{-3t}u(-t)$\n$X(s) = \\frac{5}{(s-2)(s+3)}$ (sin ROC com√∫n)")
ax2.grid(True, alpha=0.3)
ax2.legend(loc='best')
plt.show()

# ============================================
# SE√ëAL 3: e^{-a|t|} (con a=1 para ejemplo)
# ============================================
print("\n" + "="*60)
print("SE√ëAL 3: e^{-a|t|} (con a=1)")
print("="*60)

a = 1  # Valor ejemplo
poles3 = [a, -a]  # s = 1 y s = -1
zeros3 = []  # Numerador constante -2a

print(f"Polos: {poles3}")
print(f"Ceros: Ninguno finito")
print(f"ROC: -{a} < Re(s) < {a}")

# Diagrama de polos y ceros con ROC
plot_pole_zero(poles3, zeros3,
               title=f"Se√±al 3: $e^{{-{a}|t|}}$\n$X(s) = \\frac{{-2{a}}}{{s^2-{a}^2}}$",
               roc_left=-a, roc_right=a)

# ============================================
# SE√ëAL 4: e^{-2t}[u(t) - u(t-5)]
# ============================================
print("\n" + "="*60)
print("SE√ëAL 4: e^{-2t}[u(t) - u(t-5)]")
print("="*60)

# Polos y ceros
poles4 = [-2]

# Ceros: 1 - e^{-10}e^{-5s} = 0 ‚Üí e^{-5s} = e^{10} ‚Üí s = -2 + j(2œÄk/5)
k_values = range(-5, 6)  # 11 ceros para visualizar
zeros4 = [-2 + 1j*(2*np.pi*k/5) for k in k_values]

print(f"Polo: {poles4}")
print(f"Ceros: s = -2 + j(2œÄk/5), k ‚àà ‚Ñ§")
print(f"ROC: Re(s) > -2")

# Diagrama de polos y ceros
fig4, ax4 = plt.subplots()
ax4.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax4.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Polo
ax4.plot(poles4[0], 0, 'rx', markersize=10, label='Polo')

# Ceros
for zero in zeros4:
    ax4.plot(zero.real, zero.imag, 'bo', markersize=8, fillstyle='none',
            markeredgewidth=2, label='Cero' if zero == zeros4[0] else "")

# ROC
ax4.axvspan(-2, 3, alpha=0.2, color='green', label='ROC: Re(s) > -2')

# Configuraci√≥n
ax4.set_xlim(-4, 4)
ax4.set_ylim(-8, 8)
ax4.set_xlabel('Re(s)')
ax4.set_ylabel('Im(s)')
ax4.set_title("Se√±al 4: $e^{-2t}[u(t)-u(t-5)]$\n$X(s) = \\frac{1-e^{-10}e^{-5s}}{s+2}$")
ax4.grid(True, alpha=0.3)
ax4.legend(loc='best')
plt.show()

# ============================================
# RESUMEN EN UNA SOLA FIGURA
# ============================================
print("\n" + "="*60)
print("RESUMEN DE TODAS LAS SE√ëALES")
print("="*60)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Se√±al 1
ax = axes[0, 0]
for pole in poles1:
    ax.plot(pole, 0, 'rx', markersize=10)
for zero in zeros1:
    ax.plot(zero, 0, 'bo', markersize=10, fillstyle='none', markeredgewidth=2)
ax.axvspan(-2, 3, alpha=0.2, color='green')
ax.axhline(y=0, color='k', alpha=0.3)
ax.axvline(x=0, color='k', alpha=0.3)
ax.set_xlim(-4, 2)
ax.set_ylim(-1, 1)
ax.set_title("Se√±al 1: $e^{-2t}u(t)+e^{-3t}u(t)$")
ax.grid(True, alpha=0.3)

# Se√±al 2
ax = axes[0, 1]
for pole in poles2:
    ax.plot(pole, 0, 'rx', markersize=10)
ax.axvspan(2, 4, alpha=0.2, color='blue')
ax.axvspan(-5, -3, alpha=0.2, color='red')
ax.axhline(y=0, color='k', alpha=0.3)
ax.axvline(x=0, color='k', alpha=0.3)
ax.set_xlim(-5, 5)
ax.set_ylim(-1, 1)
ax.set_title("Se√±al 2: $e^{2t}u(t)+e^{-3t}u(-t)$")
ax.grid(True, alpha=0.3)

# Se√±al 3
ax = axes[1, 0]
for pole in poles3:
    ax.plot(pole, 0, 'rx', markersize=10)
ax.axvspan(-a, a, alpha=0.2, color='green')
ax.axhline(y=0, color='k', alpha=0.3)
ax.axvline(x=0, color='k', alpha=0.3)
ax.set_xlim(-2, 2)
ax.set_ylim(-1, 1)
ax.set_title(f"Se√±al 3: $e^{{-{a}|t|}}$")
ax.grid(True, alpha=0.3)

# Se√±al 4
ax = axes[1, 1]
ax.plot(poles4[0], 0, 'rx', markersize=10, label='Polo')
for zero in zeros4[:5]:  # Solo primeros 5 para claridad
    ax.plot(zero.real, zero.imag, 'bo', markersize=8, fillstyle='none',
           markeredgewidth=2, label='Cero' if zero == zeros4[0] else "")
ax.axvspan(-2, 2, alpha=0.2, color='green')
ax.axhline(y=0, color='k', alpha=0.3)
ax.axvline(x=0, color='k', alpha=0.3)
ax.set_xlim(-3, 1)
ax.set_ylim(-3, 3)
ax.set_title("Se√±al 4: $e^{-2t}[u(t)-u(t-5)]$")
ax.grid(True, alpha=0.3)
ax.legend()

plt.suptitle("DIAGRAMAS DE POLOS Y CEROS - EJERCICIO 2.9", fontsize=14)
plt.tight_layout()
plt.show()
```
